### 🔋 Ejercicio 5: Drenaje de batería (Bucle while)

A diferencia del `for`, el bucle `while` se ejecuta *mientras* una condición siga siendo verdadera (True). Es muy útil cuando no sabemos de antemano cuántas vueltas dará el bucle.

**Tu misión:**
Un robot tiene una batería inicial de 100%. Cada tarea que realiza gasta un 15% de batería. El robot debe seguir trabajando mientras le quede batería suficiente para una tarea completa (es decir, mientras tenga 15% o más).

1. Escribe un bucle `while` que se ejecute mientras la `bateria` sea mayor o igual a 15.
2. Dentro del bucle, réstale 15 a la `bateria`.
3. También dentro del bucle, súmale 1 a la variable `tareas_completadas` para llevar la cuenta de cuántas cosas ha hecho el robot.

```python
bateria = 100
tareas_completadas = 0

# --- TU CÓDIGO AQUÍ ---


# ----------------------

print("El robot completó", tareas_completadas, "tareas.")
print("Batería restante:", bateria, "%")
```

```python
# 🛑 CELDA DE PRUEBAS
assert tareas_completadas == 6, f"Error: El robot debería haber completado 6 tareas, pero tu código dice {tareas_completadas}."
assert bateria == 10, f"Error: La batería final debería ser 10%, pero tu código dice {bateria}%."
print("🎉 ¡Excelente! Has gestionado la energía del robot a la perfección usando while.")
```

```python
# @title 💡 Solución: Batería y bucle while
bateria = 100
tareas_completadas = 0

while bateria >= 15:
    bateria = bateria - 15
    tareas_completadas = tareas_completadas + 1

print("El robot completó", tareas_completadas, "tareas.")
print("Batería restante:", bateria, "%")
```

### 🍿 Ejercicio 7: El mapa de asientos del cine (Matrices)

Las listas pueden contener cualquier cosa... ¡incluso otras listas! A esto lo llamamos listas bidimensionales o matrices. Imagina una cuadrícula, como los asientos de un cine.
Para acceder a un elemento, usamos dos corchetes: `matriz[fila][columna]`. *Recuerda que tanto las filas como las columnas empiezan a contar desde 0.*

**Tu misión:**
Tenemos una pequeña sala de cine de 3x3. Los `0` son asientos libres y los `1` son asientos ocupados.
1. Queremos saber si el asiento de la **fila 1, columna 2** está ocupado. Guarda su valor en una variable llamada `estado_asiento`.
2. Ha llegado un nuevo cliente. Reserva (cambia el `0` por un `1`) el asiento ubicado en la **fila 2, columna 0**.

```python
# Nuestro cine: 3 filas y 3 columnas
cine = [
    [1, 0, 1], # Fila 0
    [0, 0, 1], # Fila 1
    [0, 0, 0]  # Fila 2
]

# --- TU CÓDIGO AQUÍ ---
# 1. Lee el estado del asiento (fila 1, columna 2)

# 2. Reserva el asiento de la fila 2, columna 0

# ----------------------

print("El estado del asiento consultado es:", estado_asiento)
print("Mapa del cine actualizado:")
for fila in cine:
    print(fila) # Imprimimos fila a fila para que se vea como una cuadrícula
```

```python
# 🛑 CELDA DE PRUEBAS
assert 'estado_asiento' in globals(), "Error: No has creado la variable 'estado_asiento'."
assert estado_asiento == 1, f"Error: Has mirado en el asiento equivocado. El valor leído fue {estado_asiento} y debería ser 1."
assert cine[2][0] == 1, "Error: No has reservado el asiento correcto. Comprueba la fila 2, columna 0."
assert sum(sum(fila) for fila in cine) == 4, "Error: Has modificado asientos extraños, debería haber exactamente 4 asientos ocupados en total."
print("🎉 ¡Perfecto! Ya sabes cómo navegar por matrices y mapas 2D.")
```

```python
# @title 💡 Solución: Mapa del cine
cine = [
    [1, 0, 1],
    [0, 0, 1],
    [0, 0, 0]
]

# 1. Leer estado
estado_asiento = cine[1][2]

# 2. Reservar asiento
cine[2][0] = 1

print("El estado del asiento consultado es:", estado_asiento)
print("Mapa del cine actualizado:")
for fila in cine:
    print(fila)
```
